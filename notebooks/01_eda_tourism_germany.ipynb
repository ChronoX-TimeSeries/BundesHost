```python
import pandas as pd
import matplotlib.pyplot as plt

# Load processed tourism dataset
df = pd.read_csv("../data/processed/tourism_long.csv")

# Convert date column to datetime
df["date"] = pd.to_datetime(df["date"])

# Sort values to ensure correct chronological order
df = df.sort_values(["state", "date"]).reset_index(drop=True)

# Quick sanity check
df.head()
```

```python
# Get unique states
states = df["state"].unique()

# Create subplot grid (8 rows x 2 columns)
fig, axes = plt.subplots(nrows=8, ncols=2, figsize=(18, 28))
axes = axes.flatten()

# Plot arrivals for each state
for i, state in enumerate(states):
    state_data = df[df["state"] == state]
    
    axes[i].plot(state_data["date"], state_data["arrivals"])
    axes[i].set_title(state)
    axes[i].set_xlabel("Date")
    axes[i].set_ylabel("Arrivals")

plt.tight_layout()
plt.show()
```

```python
# Create subplot grid (8 rows x 2 columns)
fig, axes = plt.subplots(nrows=8, ncols=2, figsize=(18, 28))
axes = axes.flatten()

# Plot overnights for each state
for i, state in enumerate(states):
    state_data = df[df["state"] == state]
    
    axes[i].plot(state_data["date"], state_data["overnight"])
    axes[i].set_title(state)
    axes[i].set_xlabel("Date")
    axes[i].set_ylabel("Overnights")

plt.tight_layout()
plt.show()
```
